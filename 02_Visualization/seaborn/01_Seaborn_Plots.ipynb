```python
sns.barplot(data=df,x="dr",y="kilo")
```

```python
sns.barplot(data=df,x="etnik",y="boy",ci="sd",estimator=np.min)
 
```

```python
sns.histplot(data=df,x="dr",kde=True, 
             edgecolor="purple",lw=2.5,ls="--",color="orange",alpha=0.8 )
plt.grid(True,color="grey",linestyle="--",alpha=0.3) # ızgara çizgilerini ekleme
```

```python
sns.rugplot(data=df,x="yas",height=0.5)
```

```python
sns.kdeplot(data=df,x="yas")
```

```python
sns.displot(data=df, x="yas", kde=True, rug=True, bins=15)
```

```python
sns.kdeplot(data=df, x="yas", bw_adjust=0.1, shade=True)
plt.grid(True, color="grey", linestyle="--", alpha=0.3)
```

```python
sns.boxplot(data=df,x="etnik",y="kilo")
```

```python
sns.boxplot(data=df,y="etnik",x="kilo",orient="h")
```

```python
sns.boxplot(data=df, x="etnik", y="kilo", hue="cinsiyet", width=0.6,palette="colorblind")
plt.legend(bbox_to_anchor=(1,0.5)) # bu kadın erkek yazan kutucuğu dışarı almamızı sağladı
```

```python
sns.violinplot(data=df,y="dr",x="yas")
```

```python
sns.violinplot(data=df,x="etnik",y="dr")
```

```python
sns.violinplot(data=df,x="etnik",y="dr",
               hue="cinsiyet",split=True,inner="quartile")
plt.legend(bbox_to_anchor=(1,0.5))
plt.grid(True,color="grey",linestyle="--",alpha=0.3)
```

```python
sns.violinplot(data=df,x="etnik",y="dr",
               hue="cinsiyet",split=True,inner="quartile",bw=0.2)
plt.legend(bbox_to_anchor=(1,0.5))
plt.grid(True,color="grey",linestyle="--",alpha=0.3)
```
