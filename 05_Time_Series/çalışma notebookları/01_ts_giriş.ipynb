**Zaman ve Tarih İşlemleri — Notlar ve Düzeltmeler**

- `date.fromisoformat()` ile verilen dizgi ISO biçiminde olmalıdır: `YYYY-MM-DD`. Ay ve gün iki basamaklı olmalıdır (ör. `2025-12-06`).
- Import seçenekleri:
  - `from datetime import date, datetime, timedelta` kullanırsanız `date(...)`, `datetime(...)`, `timedelta(...)` şeklinde çağırın.
  - `import datetime` kullanırsanız `datetime.date(...)`, `datetime.datetime(...)`, `datetime.timedelta(...)` şeklinde çağırın.
- Yaygın hatalar:
  - `AttributeError` genelde `from datetime import datetime` sonrası `datetime.date(...)` çağrılmasından kaynaklanır.
  - `ValueError` genelde `fromisoformat()`'a hatalı biçimde (eksik sıfır) verilen tarih yüzündendir.

Aşağıda yaygın `strftime`/`strptime` format dizileri:

| Format string | Anlamı |
|---------------|--------|
| `%Y`          | Yılı 4 basamak olarak gösterir (ör. 2025) |
| `%y`          | Yılı 2 basamak olarak gösterir (ör. 25) |
| `%m`          | Ayı sayı olarak gösterir (01-12) |
| `%d`          | Günü gösterir (01-31) |
| `%H`          | Saati 2 basamak olarak gösterir (00-23) |
| `%M`          | Dakikayı 2 basamak olarak gösterir (00-59) |

```python
d1 = date.fromisoformat('2025-12-06')
d2 = date(2025, 12, 6)
print(d1.strftime('%A %d. %B %Y'))
```
