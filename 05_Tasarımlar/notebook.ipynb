```python
from IPython.display import display, HTML

# Jupyter Notebook İçin Geliştirilmiş Mini Oyun Portalı
space_arcade_code = """
<div id="arcade-container" style="background: #020617; border: 2px solid #1e293b; border-radius: 16px; padding: 25px; margin: 20px auto; width: 650px; font-family: sans-serif; box-shadow: 0 20px 25px -5px rgba(0,0,0,0.5); text-align: center; color: #f1f5f9;">
    
    <div style="margin-bottom: 15px;">
        <span style="background: #1e1b4b; color: #818cf8; font-size: 11px; font-weight: bold; padding: 4px 10px; border-radius: 20px; letter-spacing: 1.5px; text-transform: uppercase;">Jupyter Engine Core v1.0</span>
        <h2 style="margin: 8px 0 2px 0; color: #ffffff; font-size: 22px; font-weight: 800;">🌌 ARCADE PORTAL: SPACE EVADER</h2>
        <p style="margin: 0; color: #64748b; font-size: 13px;">Yön tuşlarını (Sol / Sağ) kullanarak asteroidlerden kaç!</p>
    </div>

    <div style="background: #0f172a; border: 1px solid #1e293b; border-radius: 8px; padding: 10px 20px; margin-bottom: 15px; display: flex; justify-content: space-between; align-items: center;">
        <span style="color: #94a3b8; font-size: 13px; font-weight: bold;">SCORE: <span id="game-score" style="color: #38bdf8; font-size: 16px; font-family: monospace;">0</span></span>
        <span style="color: #94a3b8; font-size: 13px; font-weight: bold;">HI-SCORE: <span id="game-hiscore" style="color: #f59e0b; font-size: 16px; font-family: monospace;">0</span></span>
        <button id="start-btn" style="background: #2563eb; color: white; border: none; padding: 6px 16px; border-radius: 6px; font-weight: bold; cursor: pointer; font-size: 13px;" onclick="initGame()">OYUNU BAŞLAT</button>
    </div>

    <canvas id="spaceCanvas" width="600" height="300" style="background: #090d16; border: 1px solid #334155; border-radius: 8px; display: block; margin: 0 auto; cursor: crosshair;"></canvas>

    <div style="margin-top: 15px; font-size: 11px; color: #475569; font-family: monospace;">
        &gt;&gt; Standalone JavaScript Engine injected via IPython Renderer.
    </div>
</div>

<script>
    const canvas = document.getElementById('spaceCanvas');
    const ctx = canvas.getContext('2d');
    const scoreElement = document.getElementById('game-score');
    const hiScoreElement = document.getElementById('game-hiscore');
    
    let score = 0;
    let hiScore = 0;
    let gameActive = false;
    let player = { x: 300, y: 260, size: 20, speed: 8 };
    let obstacles = [];
    let keys = {};
    let animationId;

    // Klavye Dinleyicileri (Notebook sayfasının kaymasını engellemek için preventDefault ekli)
    window.addEventListener('keydown', function(e) {
        if([37, 39, 32].indexOf(e.keyCode) > -1) {
            e.preventDefault();
        }
        keys[e.keyCode] = true;
    });
    window.addEventListener('keyup', function(e) {
        keys[e.keyCode] = false;
    });

    function initGame() {
        if(gameActive) return;
        score = 0;
        scoreElement.innerText = score;
        obstacles = [];
        player.x = 300;
        gameActive = true;
        document.getElementById('start-btn').innerText = "YENİDEN BAŞLAT";
        
        if(animationId) cancelAnimationFrame(animationId);
        gameLoop();
    }

    function spawnObstacle() {
        if(Math.random() < 0.04) {
            obstacles.push({
                x: Math.random() * (canvas.width - 20),
                y: -20,
                size: Math.random() * 15 + 10,
                speed: Math.random() * 3 + 3
            });
        }
    }

    function update() {
        // Oyuncu Hareketi
        if (keys[37] && player.x > 0) player.x -= player.speed; // Sol
        if (keys[39] && player.x < canvas.width - player.size) player.x += player.speed; // Sağ

        // Engellerin Güncellenmesi
        for(let i = obstacles.length - 1; i >= 0; i--) {
            obstacles[i].y += obstacles[i].speed;

            // Çarpışma Testi (AABB Çakışma Algoritması)
            if (player.x < obstacles[i].x + obstacles[i].size &&
                player.x + player.size > obstacles[i].x &&
                player.y < obstacles[i].y + obstacles[i].size &&
                player.y + player.size > obstacles[i].y) {
                    gameOver();
                    return;
            }

            // Ekrandan çıkan engelleri sil ve skor ekle
            if(obstacles[i].y > canvas.height) {
                obstacles.splice(i, 1);
                score += 10;
                scoreElement.innerText = score;
                if(score > hiScore) {
                    hiScore = score;
                    hiScoreElement.innerText = hiScore;
                }
            }
        }
        spawnObstacle();
    }

    function draw() {
        // Ekran Temizleme (Uzay Arka Planı)
        ctx.fillStyle = '#090d16';
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        // Rastgele Yıldız Tozu Efekti (Statik)
        ctx.fillStyle = 'rgba(255,255,255,0.3)';
        for(let i=0; i<20; i++) {
            ctx.fillRect((i*37)%canvas.width, (i*23)%canvas.height, 2, 2);
        }

        // Oyuncu Gemisi Çizimi (Neon Mavi Üçgen)
        ctx.fillStyle = '#38bdf8';
        ctx.shadowBlur = 10;
        ctx.shadowColor = '#38bdf8';
        ctx.beginPath();
        ctx.moveTo(player.x + player.size / 2, player.y);
        ctx.lineTo(player.x, player.y + player.size);
        ctx.lineTo(player.x + player.size, player.y + player.size);
        ctx.closePath();
        ctx.fill();

        // Asteroidlerin Çizimi (Neon Kırmızı Daireler)
        ctx.fillStyle = '#f43f5e';
        ctx.shadowColor = '#f43f5e';
        obstacles.forEach(obs => {
            ctx.beginPath();
            ctx.arc(obs.x + obs.size/2, obs.y + obs.size/2, obs.size/2, 0, Math.PI * 2);
            ctx.fill();
        });
        
        // Gölge temizleme (Performans için)
        ctx.shadowBlur = 0;
    }

    function gameOver() {
        gameActive = false;
        ctx.fillStyle = 'rgba(2, 6, 23, 0.8)';
        ctx.fillRect(0, 0, canvas.width, canvas.height);
        
        ctx.fillStyle = '#ef4444';
        ctx.font = 'bold 24px sans-serif';
        ctx.textAlign = 'center';
        ctx.fillText('GAME OVER', canvas.width / 2, canvas.height / 2 - 10);
        
        ctx.fillStyle = '#94a3b8';
        ctx.font = '14px sans-serif';
        ctx.fillText('Skorunuz: ' + score + ' | Yeniden başlamak için butona basın.', canvas.width / 2, canvas.height / 2 + 20);
    }

    function gameLoop() {
        if(!gameActive) return;
        update();
        draw();
        animationId = requestAnimationFrame(gameLoop);
    }

    // İlk Çizim (Boş Ekran)
    ctx.fillStyle = '#090d16';
    ctx.fillRect(0, 0, canvas.width, canvas.height);
    ctx.fillStyle = '#64748b';
    ctx.font = '14px sans-serif';
    ctx.textAlign = 'center';
    ctx.fillText('Oyunu başlatmak için yukarıdaki butona tıklayın.', canvas.width / 2, canvas.height / 2);
</script>
"""

# HTML ve JavaScript'i Jupyter hücresine enjekte et
display(HTML(space_arcade_code))
```
