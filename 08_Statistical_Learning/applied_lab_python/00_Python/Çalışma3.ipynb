```python
def sayi_al():
    while True:
        try:
            sayi=int(input("sayi girin:"))
        except ValueError:
            print("lütfen geçerli bir sayi girin")
            continue
        else:
            print(f"Girilen sayi: {sayi}")
            return sayi
        finally:
            print("Deneme tamamlandı")
```

```python
def faktoriyel(n):
    sonuc=1
    for i in range(1, n+1):
        sonuc *=i
    return sonuc
faktoriyel(5)
```
