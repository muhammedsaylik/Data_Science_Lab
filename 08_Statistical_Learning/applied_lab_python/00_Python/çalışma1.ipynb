## 🧩 Python int Metodları ve Fonksiyonları

| Metot / Fonksiyon             | Açıklama                                                    | Örnek                                  |
| ----------------------------- | ----------------------------------------------------------- | -------------------------------------- |
| `bit_length()`                | Integer’ın binary gösteriminde kaç bit gerektiğini döndürür | `(12345).bit_length()` → `14`          |
| `to_bytes(length, byteorder)` | Integer’ı byte dizisine çevirir                             | `(12345).to_bytes(2, 'big')` → `b'09'` |
| `abs()`                       | Mutlak değer alır (built-in)                                | `abs(-678)` → `678`                    |
| `int(x)`                      | Başka tipleri integer’a çevirir                             | `int("42")` → `42`                     |
| `float(x)`                    | Integer’ı float’a çevirir                                   | `float(5)` → `5.0`                     |
| `str(x)`                      | Integer’ı string’e çevirir                                  | `str(123)` → `'123'`                   |
| `bool(x)`                     | 0 ise False, diğerleri True                                 | `bool(0)` → `False`                    |
| `bin(x)`                      | Binary string’e çevirir                                     | `bin(24)` → `'0b11000'`                |
| `oct(x)`                      | Octal string’e çevirir                                      | `oct(24)` → `'0o30'`                   |
| `hex(x)`                      | Hexadecimal string’e çevirir                                | `hex(24)` → `'0x18'`                   |


```python
sayi = 12345
negatif = -678
```

```python
print(sayi.bit_length())
```

```python
print(sayi.to_bytes(2, 'big'))

print(sayi.to_bytes(2, 'little'))
```

```python
print(abs(negatif))
```

```python
print(int("42"))       # String → integer
 
print(float(5))        # Integer → float
 
print(str(123))        # Integer → string
 
print(bool(0))          # 0 → False

print(bool(42))           # 0 olmayan → True
```

```python
print(bin(sayi))    
print(oct(sayi))    
print(hex(sayi))    
```

```python
print(sayi + 100)      
print(sayi - 200)      
print(sayi * 2)       
print(sayi // 100)     
print(sayi % 100)     
print(sayi ** 2)       
```

### Set Metodları

| Metot             | Açıklama                         | Örnek                       |
| ----------------- | -------------------------------- | --------------------------- |
| `add(x)`          | Eleman ekler                     | `kume.add(5)`               |
| `remove(x)`       | Elemanı siler (yoksa hata verir) | `kume.remove(3)`            |
| `discard(x)`      | Elemanı siler, yoksa hata vermez | `kume.discard(6)`           |
| `pop()`           | Rastgele eleman çıkarır          | `kume.pop()`                |
| `clear()`         | Küme boşaltılır                  | `kume.clear()`              |
| `union(s)`        | İki kümenin birleşimi            | `kume.union({5,6})`         |
| `intersection(s)` | Kesişim                          | `kume.intersection({2,5})`  |
| `difference(s)`   | Fark                             | `kume.difference({1,2})`    |
| `issubset(s)`     | Alt küme mi?                     | `{1,2}.issubset({1,2,3})`   |
| `issuperset(s)`   | Üst küme mi?                     | `{1,2,3}.issuperset({1,2})` |


```python
küme = {1,2,3,4,5,6}
print(küme)
```

```python
küme.add(7)
print(küme)
```

```python
küme.remove(7) # elaman yoksa hata verir
print(küme)
```

```python
küme.discard(10) # silinmek istenen eleman yoksa hata vermez
print(küme)
```

```python
küme.pop() # rastgele bir elemanı siler
print(küme)
```

```python
gecici_küme = küme.copy()
gecici_küme.clear()
print(gecici_küme)
```

```python
kume1 = {1, 2, 3}
kume2 = {3, 4, 5}
```

```python
kume1.union(kume2) # iki kümenin birleşimi
```

```python
kume1.intersection(kume2) # iki kümenin kesişimi
```

```python
{1,2}.issubset(kume1) # {1,2} kümesi kume1 in alt kümesimi 
```

```python
kume1.issuperset({1,2})  # kume1 kümesi {1,2} kümesinin üst kümesimi
```

## Koşullu İfadeler (If-Elif-Else)

```python
sayi = -10

if sayi > 0:
    print(f"{sayi} sayisi pozitiftir")
elif sayi == 0:
    print(f"{sayi} sayisi nötrdür")
else:
    print(f"{sayi} sayisi negatiftir")
```

#### Mantıksal Operatörler

```python
x = 15
y = 5

if x >10 and y < 10:
    print(f"{x} sayisi 10 dan büyüktür ve {y} sayisi 10 dan küçüktür")
```

```python
if x > 20 or y < 10:
    print(f"{x} büyük veya {y} küçük")
if not y > 10:
    print(f"{y} 10’dan büyük değil")
```

```python
sayi = -7

durum = "pozitif" if sayi >0 else "negatif veya nötr"
print(durum)
```

### while döngüsü

```python
hiz = 100
durum = True

while durum:
    if hiz <= 0:
        durum = False
        print("araba durdu")
    else:
        print(f"araba {hiz} hızla gidiyor")
        hiz -= 20

```

```python
import random

sinav_notu = random.randint(0, 100)

if sinav_notu >= 90:
    print("notunuz: AA")
elif sinav_notu >= 80:
    print("notunuz: BA")
elif sinav_notu >= 70:
    print("notunuz: BB")
elif sinav_notu >= 60:
    print("notunuz: CB")
elif sinav_notu >= 50:
    print("notunuz : CC")
else:
    print(f"sinav notunuz {sinav_notu} kaldın aq")
```
