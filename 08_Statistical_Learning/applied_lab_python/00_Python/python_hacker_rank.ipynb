```python
def print_formated(number):
    width = len(bin(number)[2:])

    for i in range(1, number +1):
        dec = str(i)
        octal = oct(i)[2:]
        hex_ = hex(i)[2:].upper() # hex i yi 16 lık tabana çeviriyor bu harf olarak döndüğü için o harflari büyütüyoruz
        binary = bin(i)[2:]    # binariy 0,1 e çevir
        print(dec.rjust(width), octal.rjust(width), hex_.rjust(width), binary.rjust(width)) # .rjust sağa yaslamamızı sağlıyor

print_formated(15)
```

```python
if __name__ == '__main__':
    records = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name, score])





all_scores = [x[1] for x in records]

unique_scores = sorted(list(set(all_scores)))

target_score = unique_scores[1]

found_names = []

for student in records:
    if student[1] == target_score:
        found_names.append(student[0])
        
found_names.sort()

for name in found_names:
    print(name)

 
```

```python
def veri_ozeti(liste):
    len(liste)
    if len(liste) > 0:
        ortalama = sum(liste) / len(liste)
        print(f"Not ortalaması: {ortalama}\n")
def not_ekle(liste):
     secim = input("öğrenci sıra no: ")
     not_degeri = int(input("Eklenecek notu giriniz: "))
     if secim == '1':
         liste.append(not_degeri)
     elif secim == '2':
         sira = int(input("hangi sira numarasina eklensin: "))
         liste.insert(sira, not_degeri)
         print(f"{not_degeri} notu {sira}.siraya eklendi")
     else:
         print("hatali seçim")
def ver_temizleme(liste):
    secim = input("lütfen silinmesini istediğiniz notun sirasini giriniz")
    if secim == '1':
        silinecek_deger = int(input("Silinecek Not Değeri: "))
        if silinecek_deger in liste:
            liste.remove(silinecek_deger)
            print(f"{silinecek_deger} listeden silindi")
        else:
            print('aradiğiniz deger bulunamadi')
    elif secim == '2':
        silinecek_index = int(input("silinecek degerin sıra numarasını giriniz: "))
        if 0 <= silinecek_index < len(liste):
            silinen = liste.pop(silinecek_index)
            print(f"{silinecek_index}. sıradaki not ({silinen}) atıldı.")
        else:
            print("Hatalı sıra numarası.")
def veriyi_duzenle(liste):
    """Listeyi sıralar veya ters çevirir (sort ve reverse örneği)."""
    print("1. Küçükten Büyüğe Sırala (sort)")
    print("2. Listeyi Ters Çevir (reverse)")
    secim = input("Seçiminiz (1/2): ")
    
    if secim == '1':
        liste.sort()
        print("hello Liste sıralandı.")
    elif secim == '2':
        liste.reverse()
        print(" Liste ters çevrildi.")

# --- ANA PROGRAM (MAIN LOOP) ---
def main():
    # Başlangıç verisi (Kirli veri simülasyonu)
    sinif_notlari = [45, 90, -10, 65, 100, 45, 30] 
    
    print(" İstatistik Veri Temizleme Aracına Hoşgeldiniz")
    
    while True: # Sonsuz döngü, kullanıcı 'q' diyene kadar döner
        print("\n--- İŞLEM MENÜSÜ ---")
        print("1. Veri Ekle (Append/Insert)")
        print("2. Veri Temizle (Remove/Pop)")
        print("3. Düzenle (Sort/Reverse)")
        print("4. Durumu Göster")
        print("q. Çıkış")
        
        komut = input("Ne yapmak istersiniz? ")
        
        if komut == '1':
            not_ekle(sinif_notlari)
        elif komut == '2':
            veri_temizle(sinif_notlari)
        elif komut == '3':
            veriyi_duzenle(sinif_notlari)
        elif komut == '4':
            veri_ozeti(sinif_notlari)
        elif komut.lower() == 'q':
            print("Çıkış yapılıyor... Veriler kaydedildi (Simülasyon).")
            break
        else:
            print("Geçersiz komut!")

# Programı başlat
if __name__ == '__main__':
    main()

     
```
