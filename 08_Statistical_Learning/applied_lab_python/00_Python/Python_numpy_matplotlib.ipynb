```python
print(np.mean((y- y.mean())**2)) # varyans formülünden ortalama kullanarak varyans bulma
```

```python
print(np.std(y))
```

```python
A = np.array(np.arange(16)).reshape((4, 4))
A
```

```python
A[1, 2] # sıralama sıfırdan başlıyo [satır, sütun] 1- 2. satır , 2-3.sütun [1, 2] = [2. satır, 3.sütun]
```

```python
A[[1, 3]] # bu şekilde birden fazla satır çağrılabilir
```

```python
A[:, [0, 2]] # bu şekilde sütun çağırabiliriz 
```

```python
A[[1, 3], [0, 2]]
```

```python
np.array([A[1, 0], A[3, 2]])
```

```python
A[[1, 3]][:, [0, 2]]
```

```python
idx = np.ix_([1, 3], [0, 2, 3])
A[idx]
```

```python
A[1:4:2, 0:3:2]
```

```python
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
keep_rows[[1, 3]] = True
keep_rows
```

```python
np.all(keep_rows == np.array([0, 1, 0, 1]))
```

```python
A[keep_rows]
```

```python
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

```python
idx_mixed = np.ix_([1, 3], keep_cols)
A[idx_mixed]
```

# Grafikler

```python
from matplotlib.pyplot import subplots
fig, ax = subplots(figsize=(8,8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)

ax.plot(x, y)
```

```python
output = subplots(figsize=(8, 8))
fig = output[0]
ax = output[1]
```

```python
fig, ax = subplots(figsize=(8, 8))
ax.plot(x, y, 'o')
```

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(x, y, marker='o')
ax.set_xlabel('x ekseni')
ax.set_ylabel('y ekseni')
ax.set_title('x vs y')
```

```python
fig.set_size_inches(12, 3)
fig
```

```python
fig, axes = subplots(nrows=2,
                     ncols=3,
                    figsize=(15, 5) )
```

```python
axes[0, 1].plot(x, y, 'o')
axes[1, 2].scatter(x, y, marker = '+', color="#1417c8ef")
axes[1, 2].set_title('x vs y (scatter)')
fig
```

```python
fig.savefig("Figure.png", dpi=400)
```

```python
fig, ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1/(1 + x**2))
ax.contour(x, y, f) # üç boyutlu verileri temsil etmek için kontur grafiği oluşturur
```

```python
fig, ax = subplots(figsize=(8, 8))
ax.contour(x, y, f, levels = 45)
```

```python
fig, ax = subplots(figsize=(8, 8))
ax.imshow(f)
```
