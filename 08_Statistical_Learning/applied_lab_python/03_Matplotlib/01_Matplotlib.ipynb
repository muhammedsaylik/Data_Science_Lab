```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic') # estetik açıdan grafikler için farklı stiller seçmeni sağlıyor
# seaborn karşılığı = seaborn.set_style()  -> - Parametreleri: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'.
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

```python
import seaborn as sns
sns.set_style('whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
sns.set_style('darkgrid')
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))

plt.grid(axis='both', color = 'black' , linestyle='--', alpha = 0.5, linewidth=0.9) # gridçizgilerinin parametreleri , linewidth çizgi kalınlığını ayarlıyor
```

```python
sns.set_style('whitegrid')
plt.plot(x, np.sin(x -0))
plt.plot(x, np.sin(x -1))
plt.plot(x, np.sin(x -2))
plt.plot(x, np.sin(x -3))
plt.plot(x, np.sin(x -4))
plt.plot(x, np.sin(x -5))
plt.grid(axis='both', color='gray', linestyle='--')
```

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5) # bu fonksiyonlar kısaca görünür eksen aralıklarını belirlememizi sağlıyor
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))

plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))

plt.axis('tight') # 'tight' sadece dar açıdan görsele bakmamızı sağlıyor görseli grafiğin bulunduğu yere kadar daraltıyor
```

```python
plt.plot(x, np.sin(x))

plt.axis('equal') # eksenlerin ölçeklerini eşitler 
```

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve') # genel başlık eklemek için kullanılır
plt.xlabel('x') # x eksenin başlığını eklemek için kullanılır
plt.ylabel('sin(x)') # y ekseninin başlığını eklemek için kullanılır 
```

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')

plt.axis('equal')
plt.legend(frameon=False, loc = 'upper right') # frameon legendlerin kutu içinde olup olmamasını ayarlar loc konum ayarlaman için 
```

### Simple Scatter Plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
    label="marker='{0}'".format(marker))
```

```python
plt.plot(x, y, '-ok')
```

```python
plt.plot(x, y, '-p', color='gray',
markersize=15, linewidth=4, # markersize marker boyutu, 
markerfacecolor='white', # marker iç dolgu rengi 
markeredgecolor='gray', # marker kenar rengi
markeredgewidth=2) # marker kenar kalınlığı
plt.ylim(-1.2, 1.2)
```

```python
plt.scatter(x, y, marker = 'o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, marker='o', cmap='viridis', c = colors, s = sizes, alpha=0.5)
```
