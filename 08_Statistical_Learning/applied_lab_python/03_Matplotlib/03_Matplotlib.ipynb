## plt.axes: Subplots by Hand

```python
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], xticks=[], ylim=[-1.2, 1.2])
ax2 = fig.add_axes([0.1, 0.1, 0.8,0.4], ylim=[-1.2, 1.2])
x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

## Arrows and Annotation

```python
fig, ax = plt.subplots(figsize=(12, 6))
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x), linewidth=2.5, label='cos(x)')

# Annotate local maximum at (2π, 1)
ax.annotate('local max', xy=(2*np.pi, 1), xytext=(2*np.pi - 1.5, 1.4),
            arrowprops=dict(arrowstyle='->', color='red', lw=2.5),
            fontsize=11, color='red', weight='bold',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.3))

# Annotate local minimum at (π, -1)
ax.annotate('local min', xy=(np.pi, -1), xytext=(np.pi + 1, -1.4),
            arrowprops=dict(arrowstyle='->', color='blue', lw=2.5),
            fontsize=11, color='blue', weight='bold',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='lightblue', alpha=0.3))

ax.grid(True, alpha=0.4, linestyle='--')
ax.set_xlabel('x', fontsize=12, weight='bold')
ax.set_ylabel('cos(x)', fontsize=12, weight='bold')
ax.set_title('Cosine Function with Annotations', fontsize=14, weight='bold')
ax.legend(fontsize=11)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
```

## Hiding Ticks or Labels

```python
ax = plt.axes()
ax.plot(np.random.rand(50))
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi)
```
