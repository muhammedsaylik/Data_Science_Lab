### Hierarchical Indexing

```python
import numpy as np
import pandas as pd
```

```python
index = [('California', 2000), ('California', 2010),('New York', 2000), ('New York', 2010), ('Texas', 2000), ('Texas', 2010)] 
populations = [33871648, 37253956, 18976457, 19378102, 20851820, 25145561]

pop = pd.Series(populations, index  = index)
pop
```

```python
pop[('California', 2010):('Texas', 2000)]
```

```python
pop[[i for i in pop.index if i[1]==2010]] # böyle istenen belirli kısımlar alınabilir ama büyük veri setleri için mantıklı değil
```

```python
# çoklu indexleme

index = pd.MultiIndex.from_tuples(index)
index
```

```python
pop = pop.reindex(index)
pop
```

```python
pop[:, 2010]
```

```python
pop_df = pop.unstack() # unstack() çok indexli bir yapıyı DataFrame haline getirmeye yarar
pop_df
```

```python
pop_df = pd.DataFrame({'total':pop,
                       'under18': [9267089, 9284094,
4687374, 4318033,
5906301, 6879014]})
```

```python
pop_df
```

- unstack() sadece DataFrame üzerinde veya Series üzerinde çalışıyor bir reshape metodu ve çok indexli verilerde kullanılıyor aşağdaki gibi indexsi sütun haline getiriyor

```python
f_u18 = pop_df['under18'] / pop_df['total']
f_u18.unstack()
```
