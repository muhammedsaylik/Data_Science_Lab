```python
# GrupBy ve agregasyon

df = pd.DataFrame({
    'Departman': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT'],
    'Ad': ['Ali', 'Betül', 'Cemal', 'Deniz', 'Elif', 'Fatih'],
    'Maaş': [3000, 4000, 5000, 3500, 4500, 5500],
    'Yaş': [25, 30, 35, 28, 32, 26]
})
df
```

```python
# departman bazında gruplama
gruplar = df.groupby('Departman')
gruplar
```

```python
for ad, grup in gruplar:
    print(ad)
    print(grup)
```

```python
len(gruplar)
```

```python
# grup bilgisi
gruplar.groups
```

```python
# agregasyon fonksiyonları
print(df.groupby('Departman')['Maaş'].max())
print(df.groupby('Departman')['Maaş'].count())
print(df.groupby('Departman')['Maaş'].std())
print(df.groupby('Departman')['Maaş'].first())
print(df.groupby('Departman')['Maaş'].last())
```

```python
# çoklu agregasyon
ozet = df.groupby('Departman')['Maaş'].agg(['sum', 'mean', 'count', 'std'])
ozet
```

```python
# string işlemler

df = pd.DataFrame({
    'İsim': ['ali', 'BETÜL', 'Cemal', 'deniz çakmak'],
    'Email': ['ali@example.com', 'betul@example.com', 'cemal@example.com', 'deniz@example.com']
})
df
```

```python
#Büyük harf
df['isim_upper'] = df['İsim'].str.upper()
df
```

```python
# küçük harf

df['İsim_lower'] = df['İsim'].str.lower()
df
```

```python
# Başharfi büyüt
df['İsim_title'] = df['İsim'].str.title()
df
```

```python
# uzunluk 
df['isim_uzunluğu'] = df['İsim'].str.len()
df
```

```python
# İçeriyor
df['Ali_içeriyor'] = df['İsim'].str.contains('ali', case=False)
print(df)

# Parça çıkar
df['Alan'] = df['Email'].str.extract(r'(.+)@', expand=False)
print(df)

# Böl (Split)
df[['Ad', 'Soyad']] = df['İsim'].str.split(expand=True)
print(df)

# Değiştir
df['İsim_Temiz'] = df['İsim'].str.replace('ç', 'c')
print(df)

# Başlangıç/bitiş kontrol
df['com_ile_bitis'] = df['Email'].str.endswith('.com')
print(df)
```

```python
df = pd.DataFrame({
    'İsim': ['Ali', 'Betül', 'Cemal', 'Deniz'],
    'Yaş': [25, 30, 35, 28],
    'Maaş': [3000, 4000, 5000, 3500]
})
```

```python
kosullar = [ df['Yaş'] < 26, (df['Yaş'] >= 26 ) & (df['Yaş'] < 31), df['Yaş'] >= 31]
secenekler = ['Çok Genç', 'Genç', 'Yaşlı']

df['Yaş'] = pd.to_numeric(df['Yaş'], errors = 'coerce')
df['Yaş_Grubu'] = np.select(kosullar, secenekler, default='Bilinmiyor')
df
```

```python
# pd.cut

df['Yaş_bin'] = pd.cut(df['Yaş'], bins=[0, 26, 31, 100], labels=['Çok Genç', 'Genç', 'Yaşlı'])
df
```

```python
# pd.qcut
df['Maaş_Quartile'] = pd.qcut(df['Maaş'], q=3, 
                              labels=['Düşük', 'Orta', 'Yüksek'])
df
```
