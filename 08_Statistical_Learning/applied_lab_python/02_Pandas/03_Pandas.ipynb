## Combining Datasets: Concat and Append

- append pandastan kaldırıldı concat kullanman yeterli

```python
import numpy as np
import pandas as pd
```

```python
def make_df(cols, ind):
    data = {c:[str(c) + str(i) for i in ind] for c in cols}
    return pd.DataFrame(data, ind)
make_df('ABC', range(3))
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

np.concatenate([x, y, z]) # numpy kullanarak dizi birleştirme 
```

```python
ser1 = pd.Series(['A', 'B', 'C'], index=[1, 2, 3])
ser2 = pd.Series(['D', 'E', 'F'], index=[4, 5, 6])
pd.concat([ser1, ser2])
```

```python
pd.concat([ser1, ser2], axis=1).fillna('x')
```

```python
df1 = make_df('AB', [1,2])
df2 = make_df('AB', [3, 4])
display(df1)
display(df2)
display(pd.concat([df1, df2])) # concat df leri birleştirir
```

```python
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
display(df3)
display(df4)
display(pd.concat([df3, df4], axis=1))
display(pd.concat([df3, df4], axis=0))
```

```python
x = make_df('AB', [0, 1])
y = make_df('AB', [2, 3])

y.index =x.index
display(x)
display(y)
display(pd.concat([x, y]))  # np.concatenate ile pd.concat arasındaki fark concat ın indexler aynı olsabile onları korumasıdır
```

```python
df5 = make_df('ABC',  [1, 2])
df6 = make_df('BCD', [3, 4])

display(df5)
display(df6)
display(pd.concat([df5, df6]))
```

### Combining Datasets: Merge and Join

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
'hire_date': [2004, 2008, 2012, 2014]})

display(df1)
display(df2)
```

```python
df3 = pd.merge(df1, df2) # merge() df leri ortak sütun veya sütunlara göre birleştirmeye yarar  JOIN türlerini destekler: inner, left, right, outer 

df3
```

```python
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'], 'supervisor': ['Carly', 'Guido', 'Steve']})
pd.merge(df3, df4)
```

```python
df5 = pd.DataFrame({'group': ['Accounting', 'Accounting','Engineering', 'Engineering', 'HR', 'HR'],
                   'skills': ['math', 'spreadsheets', 'coding', 'linux','spreadsheets', 'organization']})

pd.merge(df1, df5)
```

#### Specification of the Merge Key

```python
pd.merge(df1, df2, on='employee') # on parametresi sütun ile birleştirmeye yarar çalışan sütunundaki isimleri alıp birleştirdi
```

```python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
'salary': [70000, 80000, 120000, 90000]})

pd.merge(df1, df3, left_on="employee", right_on="name") # tam olarak anlamadım 
```

```python
pd.merge(df1, df3, left_on="employee", right_on="name").drop('name', axis=1)
```

- ek olarak bilinmesi gerekenler 'how =' parametresi
 - 'inner' ortak eşleşen satırları alır
 - 'outer' her iki dataframe de alır eşleşmeyenleri nan ile doldurur

```python
pd.merge(df1, df2, how='inner')
```

### GroupBy: Split, Apply, Combine

```python
df = pd.DataFrame({'key':['A', 'B', 'C', 'A', 'B', 'C'], 'data' : range(6)}, columns=['key', 'data'])
df
```

```python
df.groupby('key') # groupby() verileri belirli bir sütuna göre guruplayarak bu gruplar üzerinden toplulaştırma (aggregation) işlemleri yapmak için kullanılır
```

```python
df.groupby('key').sum() # burada key e göre guruplayıp data daki değerleri topluyor
```

```python
planets.groupby('method')
```

```python
planets.groupby('method')['orbital_period']  # groupby() indexleme işlemlerine izin verir
```

### Aggregate, filter, transform, apply

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key':['A', 'B', 'C', 'A', 'B', 'C'], 'data1':range(6), 'data2':rng.randint(0, 10, 6)}, columns=['key', 'data1', 'data2'])
df
```

```python
df.groupby('key').aggregate(['min', np.median, max]) # bir guruptaki bir çok satırı alır ve geriye tek bir değer veya her sütun için bir özet döndürür
```

```python
df.groupby('key').aggregate({'data1':'min', 'data2': 'max'}) 
```

```python
def filter_func(x):
    return x['data2'].std() > 4
display(df)
display(df.groupby('key').std())
display(df.groupby('key').filter(filter_func)) # gurup bazlı eleme yapmak için kullanılır
```

```python
df.groupby('key').transform(lambda x: x - x.mean()) # verinin şeklini koruyarak veriyi gurubun geneline göre değiştirmeye yarar
```

```python
def norm_by_data2(x):
    x
    x['data1'] /= x['data2'].sum()
    return x

display(df)
display(df.groupby('key').apply(norm_by_data2)) # gruba özel yazılan python kodunu uygulamak için kullanılır
```

```python
L = [0, 1, 0, 1, 2, 0]
df.groupby(L).sum()
```

```python
df.groupby(df['key']).sum()
```

```python
df2 = df.set_index('key')
mapping = {'A': 'vowel', 'B': 'consonant', 'C': 'consonant'}
df2.groupby(mapping).sum()
```

```python
df2.groupby(str.lower).mean()
```

```python
decade = 10 * (planets['year'] // 10)
decade = decade.astype(str) + 's'
decade.name = 'decade'
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)
```
