### Pivot Tables

```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')['survived'].sum()
```

```python
titanic.groupby('class')['survived'].sum()
```

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack() # kod sex ve class a göre guruplama yapıyor ve bunu survived değişkeni üzerinden aggregate komutunu kullanarak ortalama üzerinden ölçüyor
# indexlemenin doğru olması için unstack() kullanılır bu sayede satırlar sex sütunlar class olur
```

```python
titanic.pivot_table('survived', index='sex', columns='class') # üstte olan şeyin aynısını yapıyor 
# parametre ile ölçümü değiştirebiliyorsun burada ortalama kullanılmış aggfunc = '' ile sum, count, median kullanabilirsin 
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class') # burada pd.cut ile ki bu sürekli değişkenleri belirli aralıklara bölmemizi sağlar yaş değişkenini belirli aralıklara böldükten sonra diğer değişkenler ile gurupluyoruz
```

```python
fare = pd.qcut(titanic['fare'], 3)    # pd.cut tan farklı olarak burada aralıklar verinin dağılımına göre otomatik oluşur sadece kaç aralık girmek istediğini yazarsın
titanic.pivot_table('survived', ['sex', age], [fare, 'class']) # (ölçüm değişkeni, satırlar, sütunlar)
```

```python
titanic.pivot_table(index='sex', columns='class', aggfunc={'survived':sum, 'fare':'mean'})
```

```python
titanic.pivot_table('survived', index  = 'sex', columns='class', margins=True)
```
