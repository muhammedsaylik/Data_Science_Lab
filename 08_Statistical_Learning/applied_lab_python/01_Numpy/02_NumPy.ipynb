### Array Concatenation and Splitting

#### Concatenation of arrays

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])  # dizileri birleştirmek için kullanılır
```

```python
z = [99, 99, 99]

np.concatenate([x, y, z])
```

```python
grid = np.array([[1, 2, 3], [4, 5, 6]])
np.concatenate([grid, grid])
```

```python
np.concatenate([grid, grid], axis=1) # axis = 1 boyunca birleştir
```

```python
x = np.array([1, 2, 3])
grid = np.array([[9, 8, 7], [6, 5, 4]])

np.vstack([x, grid]) # diziyi dikey olarak birleştirir
```

## Splitting of arrays

```python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x,[3, 5])
print(x1, x2, x3)
```

```python
grid = np.arange(16).reshape((4, 4))
grid
```

```python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
```
