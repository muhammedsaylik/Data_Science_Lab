## Computation on NumPy Arrays: Universal Functions

```python
import numpy as np
```

```python
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values)) # np.empty boş bir dizi oluşturur
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output
values = np.random.randint(1, 10, size=5)
compute_reciprocals(values)
    
```

```python
big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

```python
print(compute_reciprocals(values))
print(1/values)
```

### tirigonometrik fonksiyonlar

```python
theta = np.linspace(0, np.pi, 3)

print(f"theta: {theta}")
print(f" sin(theta): {np.sin(theta)}")
print(f" cos(theta): {np.cos(theta)}")
print(f" tan(theta): {np.tan(theta)}")
```

```python
x = [-1, 0, 1]

print("x=",x)
print("arsin(x): ",np.arcsin(x))
print("arcos(x): ",np.arccos(x))
print("arctan(x): ",np.arctan(x))
```

```python
x = [1, 2, 3]

print(f"x   = {x}")
print(f"e^2 = {np.exp(x)}")
print(f"2^x = {np.exp2(x)}")
print(f"3^x = {np.power(3, x)}")
```

```python
x = [1, 2, 4, 10]

print("x        =", x)
print("ln(x)    =", np.log(x))
print("log2(x)  =", np.log2(x))
print("log10(x) =", np.log10(x))
```

```python
x = [0, 0.001, 0.01, 0.1]
print("exp(x) - 1 =", np.expm1(x))
print("log(1 + x) =", np.log1p(x))
```

```python
from scipy import special

x = [1, 5, 10]

print("gamma(x)     =", special.gamma(x))
print("ln|gamma(x)| =", special.gammaln(x))
print("beta(x, 2)   =", special.beta(x, 2))
```

```python
x = np.array([0, 0.3, 0.7, 1.0])
print("erf(x)  =", special.erf(x))
print("erfc(x) =", special.erfc(x))
print("erfinv(x) =", special.erfinv(x))
```

```python
x = np.arange(5)
y = np.empty(5)
np.multiply(x, 10, out=y)
print(y)
```

```python
y = np.zeros(10)
np.power(2, x, out=y[::2])
print(y)
```

```python
z = np.sin(x) ** 10 + np.cos(10 + x*y) * np.cos(x)
```
