```python
np.zeros(10, dtype=int) # zeros 0 lardan oluşan dizi veya matris oluşturmamızı sağlar
```

```python
np.ones((3, 5), dtype=int) # 1 lerden oluşan dizi veya matris oluşturmamızı sağlar
```

```python
np.full((3, 5), 3.14) # belirttiimiz elemandan oluşan bir matris oluşturmamızı sağlar
```

```python
np.arange(0, 20, 2) # 0 dan başlayıp 20 ye kadar 20 dahil olmamak üzere 2 şer artan bir dizi oluşturmamızı sağlar
```

```python
np.linspace(0, 1, 5) # 0-1 aralığında eşit aralıklı 5 değer içeren bir dizi oluştur
```

```python
np.random.random((3, 3))  # 3x3 lük 0-1 aralığında düzgün dağılıma sahip bir matris oluşur
```

```python
np.random.normal(0, 1, (3, 3)) # ortalaması 0 ve sıtandart sapması 1 olan normal dağılıma sahip rastgele değerlerden oluşan 3x3 lük matris oluştur
```

```python
np.random.seed(42) # rastgelelik için sabit seed değeri bu sayede sürekli aynı değerlere sahip matris oluşur
matris = np.random.normal(loc=0, scale=1, size=(3, 3)) # parametreleri kullanarak yine istenen matris oluşturulabilir
matris
```

```python
np.random.randint(0, 10, (3, 3)) # [0, 10) aralığında rastgele tam sayılardan oluşan 3x3 boyutunda bir dizi oluştur
```

```python
np.eye(3) # 3x3 lük birim matris oluştur
```
