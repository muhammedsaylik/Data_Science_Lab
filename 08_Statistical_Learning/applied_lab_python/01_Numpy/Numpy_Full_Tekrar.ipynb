```python
import time

def python_dongusu(n):
    sonuc = []
    for i in range(n):
        sonuc.append(i ** 2)
    return sonuc
```

```python
def numpy_vetorize(n):
    return np.arange(n) ** 2
```

```python
n = 1000000
```

```python
start = time.time()
python_dongusu(n)
python_sure = time.time() - start
```

```python
start = time.time()
numpy_vetorize(n)
numpy_sure = time.time() - start
```

```python
start = time.time()
python_dongusu(n)
numpy_sure = time.time() -start
```

```python
print(f"Python: {python_sure:.4f} saniye")
print(f"NumPy: {numpy_sure:.4f} saniye")
print(f"NumPy {python_sure/numpy_sure:.1f}x daha hızlı!")
```
